## Segment 4 - Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features
#### The functional method

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2 ,1]
plt.bar(x, y)
plt.grid()
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
```

```python
z = [1, 2, 3, 4, 0.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
```

```python
plt.pie(z, labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/Users/sometheasiek/Documents/PsychPress/Ex_Files_Python_Data_Science_EssT_Pt_1/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.grid()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
ax.set_xlabel('Car Names')
ax.set_ylabel('miles/gal')
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.grid()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')
ax.set_xlabel('Car Names')
ax.set_ylabel('miles/gal')   
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

mpg.plot()

ax.grid()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')

ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('Car Names')
ax.set_ylabel('miles/gal')   

ax.legend(loc='best')
ax.set_ylim([0, 45])
ax.annotate('Toyota Corolla', xy=(19, 33.9), xytext=(21, 35), arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Lowest Point', xy=(14, 10.4), xytext=(9, 7), arrowprops=dict(facecolor='gray', shrink=0.05))
ax.annotate('Lowest Point', xy=(15, 10.4), xytext=(17, 7), arrowprops=dict(facecolor='gray', shrink=0.05))
```
