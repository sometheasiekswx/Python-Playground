## Segment 5 - Starting with parametric methods in pandas and scipy

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-darkgrid')
```

### The Pearson Correlation

```python
address = 'Data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
X = cars[['mpg', 'wt', 'hp', 'qsec']]
sb.pairplot(X)
```

### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
wt = cars['wt']
hp = cars['hp']
qsec = cars['qsec']

pearsonr_correlation, p_value = pearsonr(mpg, hp)
print(pearsonr_correlation, '|', p_value)
```

```python
pearsonr_correlation, p_value = pearsonr(mpg, qsec)
print(pearsonr_correlation, '|', p_value)
```

```python
pearsonr_correlation, p_value = pearsonr(mpg, wt)
print(pearsonr_correlation, '|', p_value)
```

### Using pandas to calculate the Pearson correlation coefficient

```python
corr = X.corr()
corr
```

### Using Seaborn to visualize the Pearson correlation coefficient

```python
sb.heatmap(corr)
```

```python
sb.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values)
```
