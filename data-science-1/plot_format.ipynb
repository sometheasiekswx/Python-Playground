## Segment 3 - Plot formatting

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

### Defining plot color

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]

plt.bar(x, y)
```

```python
wide = [0.5, 0.25, 0.75, 0.5, 0.25, 0.1, 0.9, 0.2, 0.8]
color = ['salmon', 'crimson', 'teal']
plt.bar(x, y, width=wide, color=color)
```

```python
address = 'Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

df = cars[['cyl', 'mpg', 'wt']]
df.plot()
```

```python
color_theme = ['darkgray', 'lightsalmon', 'powderblue']
df.plot(color=color_theme)
```

```python
z = [1, 2, 3, 4, 0.5]
plt.pie(z)
plt.show()
```

```python
color_theme = ['#A9A9A9', '#000000', "#B0EB0E", "#FFFFFF", "#BDB76B"]
plt.pie(z, colors=color_theme)
plt.show()
```

### Customizing line styles

```python
x1 = range(0, 10)
y1 = range(10, 0, -1)

plt.plot(x, y, ds='steps', lw=5)
plt.plot(x1, y1, ls="--", lw= 10)
```

### Setting plot markers

```python
plt.plot(x, y, marker='x', mew=20)
plt.plot(x1, y1, marker="+", mew= 20)
```

```python
plt.plot(x, y, marker='1', mew=10)
plt.plot(x1, y1, marker="*", mew= 10)
```
